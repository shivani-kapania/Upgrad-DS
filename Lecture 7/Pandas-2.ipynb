## pd.concat()

Concatenate pandas objects along a particular axis with optional set logic along the other axes.



```python
s1 = pd.Series(['a', 'b', 'c'])
s2 = pd.Series(['d', 'e'])
```

```python
s1
```

```python
s2
```

```python
pd.concat([s1, s2], axis=0)
```

```python
s1.append(s2)
```

```python
#pd.concat([s1, s2], axis=1)
```

```python
pd.concat([s1, s2], ignore_index=True)
```

Add a hierarchical index at the outermost level

```python
hello = pd.concat([s1, s2], keys=['s1', 's2'], names=['Series', 'row'])
```

```python
hello
```

```python
hello.loc['s1']
```

```python
hello.loc['s1', 2]
```
