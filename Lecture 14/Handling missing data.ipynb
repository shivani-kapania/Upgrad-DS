Question. Find out the rows which have more than 5 missing values. 

```python
#length of a row in df
df.shape[1]-5+1 #rows with less than 5 missing values if we set this as thresh . 
```

```python
round((len(df[df.isnull().sum(axis=1) > 5]) / df.shape[0])*100,2)
```

```python
df.isnull().sum(axis=1) <= 5
```

```python
# drop the rows with atleast 5 missing values . atmost how many non-missing values they have? = length of row - 5 
# we are dropping the rows which have less than 13 non-missing values 
# because if there are more than 5 missing values -> there are less than 13 non-missing values 
df = df[df.isnull().sum(axis=1) <= 5]
```
