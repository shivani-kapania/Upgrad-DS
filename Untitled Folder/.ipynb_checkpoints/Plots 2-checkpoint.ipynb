```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend()
```

### Bar chart

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 136, 89, 27]
```

```python
plt.bar(np.arange(len(company)), revenue, label='revenue')
```

```python
plt.bar(company, revenue, label='revenue')

plt.ylabel('Revenue (Billion)')
plt.title('Big 4 Stocks')
plt.legend()
```

```python
np.arange(len(company))
```

```python
plt.bar(np.arange(len(company)), revenue, label='revenue')
plt.ylabel('Revenue (Billion)')
plt.title('Big 4 Stocks')
plt.legend()
plt.xticks(np.arange(len(company)), company)
```

```python
profit = [40, 2, 34, 12]
```

```python
plt.bar(np.arange(len(company))-0.2, revenue, width=0.4, label='revenue')
plt.bar(np.arange(len(company))+0.2, profit, width=0.4, label='profit')
plt.ylabel('Revenue (Billion)')
plt.title('Big 4 Stocks')
plt.xticks(np.arange(len(company)), company)
plt.legend()
```

```python
data = [[5., 25., 50., 20.],
  [4., 23., 51., 17.],
  [6., 22., 52., 19.]]

X = np.arange(4) #np.array([2, 6, 10, 14])
plt.bar(X + 0.00, data[0], color = 'b', width = 0.25)
plt.bar(X + 0.25, data[1], color = 'g', width = 0.25)
plt.bar(X + 0.50, data[2], color = 'r', width = 0.25)
plt.grid()
plt.show()
```

### Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar, rwidth=0.8)
```

```python
plt.hist(blood_sugar, rwidth=0.8, bins=4)
```

```python
help(plt.hist)
```

```python
plt.hist(blood_sugar, rwidth=0.95, bins=[80, 100, 125, 150], color='g')
```

```python
plt.hist(blood_sugar, rwidth=0.95, bins=[80, 100, 125, 150], color='g', orientation='horizontal')
```

```python
plt.hist(blood_sugar, rwidth=0.95, bins=[80, 100, 125, 150], color='g')
plt.xlabel('Sugar levels')
plt.ylabel('Number of patients')
plt.title('Blood sugar summary')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 128, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], rwidth=0.95, bins=[80, 100, 125, 150], label=['men', 'women'])
plt.xlabel('Sugar levels')
plt.ylabel('Number of patients')
plt.title('Blood sugar summary')
plt.legend()
```

```python
plt.hist(blood_sugar, rwidth=0.95, bins=[80, 100, 125, 150], histtype='step')
plt.xlabel('Sugar levels')
plt.ylabel('Number of patients')
plt.title('Blood sugar summary')
```

### Pie chart

```python
expenses = [1400, 600, 300, 410, 250]
expenses_labels = ['Home', 'Food', 'Internet', 'Travel', 'Convenience']
```

```python
plt.pie(expenses, labels = expenses_labels)
```

We want percentages?

```python
plt.pie(expenses, startangle = 90, labels=expenses_labels, autopct='%.2f%%', shadow=True, explode=[0, 0, 0.3, 0, 0]); # fmt%pct
```
